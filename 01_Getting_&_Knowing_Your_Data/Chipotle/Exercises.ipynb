### Step 9. Which was the most-ordered item? 

```python
# Group the data by item_name and select the quantity column
c = chipo.groupby('item_name')['quantity']

# Sum up the quantities for each unique item
c = c.sum()

# Sort the items by quantity in descending order
c = c.sort_values(ascending=False)

# Get the first row (most ordered item)
most_ordered_item = c.head(1)

most_ordered_item
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
# Use iloc[0] to access the actual number 
# and converts it to integer type for clean output
int(most_ordered_item.iloc[0])
```

### Step 11. What was the most ordered item in the choice_description column?

```python
c = chipo.groupby('choice_description')['quantity']
c = c.sum()
c = c.sort_values(ascending=False)
most_ordered_item_choice_description = c.head(5)
most_ordered_item_choice_description
```

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
chipo.item_price.dtype
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x.replace("$", "")))
chipo['item_price']
```

#### Step 13.c. Check the item price type

```python
chipo.item_price.dtype
```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue = (chipo['item_price'] * chipo['quantity']).sum()
print('Revenue was: $' + str(np.round(revenue, 2)))
```

### Step 15. How many orders were made in the period?

```python
total_orders = chipo.order_id.nunique() # count unique values in the order_id column
total_orders
```

### Step 16. What is the average revenue amount per order?

```python
# Solution 1
average_revenue = revenue / total_orders
average_revenue
print(f"${average_revenue:.2f}")
```

```python
# Solution 2
chipo['revenue'] = chipo['quantity'] * chipo['item_price']
average_order = chipo.groupby('order_id')['revenue'].sum().mean()
print(f"${average_order:.2f}")
```
