### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = df
chipo
```

### Step 4. How many products cost more than $10.00?

```python
chipo.item_price.dtype
```

```python
chipo['item_price'] = chipo['item_price'].apply(
    lambda x: float(x.replace("$", "") if isinstance(x, str) else x)
)
```

```python
chipo_filtered = chipo[chipo['quantity'] == 1]
chipo_filtered
```

```python
high_price_products = chipo_filtered[chipo_filtered['item_price'] > 10]
high_price_products
```

```python
high_price_products.item_name.nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
item_prices = chipo[['item_name', 'item_price']]
item_prices = item_prices.drop_duplicates()
item_prices = item_prices.sort_values(by='item_price', ascending=True)
item_prices
```

### Step 6. Sort by the name of the item

```python
sorted_by_name = item_prices.sort_values(by='item_name')
sorted_by_name
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
sorted_by_price = chipo.sort_values(by='item_price', ascending=False)
sorted_by_price.head(1)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie_salad_bowl = chipo[chipo['item_name'] == 'Veggie Salad Bowl']
veggie_salad_bowl
```

```python
len(veggie_salad_bowl.quantity)
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
more_than_one_canned_soda = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]
more_than_one_canned_soda
```

```python
len(more_than_one_canned_soda.quantity)
```
